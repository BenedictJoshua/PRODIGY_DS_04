```python
import pandas as pd

df = pd.read_csv("twitter_training.csv", header=None)

print(df.head())
print(df.shape)
```

```python
df.columns = ["ID", "Topic", "Sentiment", "Text"]

print(df.head())
print(df["Sentiment"].value_counts())
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df["Sentiment"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(sentiment_counts.index, sentiment_counts.values)

plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

```python
top_topics = df["Topic"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_topics.index, top_topics.values)

plt.title("Top 10 Most Discussed Topics")
plt.xlabel("Topic")
plt.ylabel("Number of Tweets")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x="Sentiment",
    order=df["Sentiment"].value_counts().index
)

plt.title("Sentiment Analysis of Social Media Data")
plt.show()
```

## Observations

1. Social media posts were analyzed based on their sentiment categories.

2. The dataset contains Positive, Negative, Neutral, and Irrelevant sentiments.

3. Sentiment distribution shows how users feel about different topics and brands.

4. Some topics received significantly more attention than others.

5. Visualizations help identify public opinion patterns and overall sentiment trends in social media discussions.
